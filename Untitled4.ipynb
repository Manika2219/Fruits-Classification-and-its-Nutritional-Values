```python
# Import necessary libraries
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import gc

# Clear any previous session to free up memory
tf.keras.backend.clear_session()

# Enable garbage collection for memory management
gc.collect()

# Set image size and batch size
img_size = 64  # Reduce image size to 64x64 to save memory
batch_size = 32  # Set batch size to 32 for efficient memory usage

# Set the correct path for the dataset (Update path according to where your dataset is located)
data_dir = '/content/fruits-360_dataset_100x100/fruits-360/Training'  # Replace with correct dataset path

# Using ImageDataGenerator for efficient image loading and augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,      # Rescale images between 0 and 1
    validation_split=0.2  # Set 20% of data for validation
)

# Load training and validation data using flow_from_directory
train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(img_size, img_size),  # Resize images
    batch_size=batch_size,
    class_mode='sparse',  # Use sparse labels
    subset='training'     # Load training data
)

validation_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(img_size, img_size),  # Resize images
    batch_size=batch_size,
    class_mode='sparse',  # Use sparse labels
    subset='validation'   # Load validation data
)

# Build the CNN model
model = models.Sequential()

# Convolutional layers
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(img_size, img_size, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

# Flatten and fully connected layers
model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(train_generator.num_classes, activation='softmax'))  # Output layer

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Display the model's architecture
model.summary()

# Train the model using the data generators
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // batch_size,
    epochs=10  # You can increase the number of epochs based on the performance
)

# Plot training and validation accuracy
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0, 1])
plt.legend(loc='lower right')

# Evaluate the model
test_loss, test_acc = model.evaluate(validation_generator, verbose=2)
print(f"Test accuracy: {test_acc}")
```

```python
nutritional_info = {
    'Apple 6': {'Calories': 52, 'Carbs': 14, 'Fiber': 2.4, 'Vitamin C': 7},  # per 100g
    'Banana 1': {'Calories': 96, 'Carbs': 27, 'Fiber': 2.6, 'Vitamin C': 8.7},
    'Orange 1': {'Calories': 43, 'Carbs': 8.3, 'Fiber': 2.2, 'Vitamin C': 53},
    # Add more fruits and nutritional values
}
```

```python
# Load and preprocess a new image for prediction
from tensorflow.keras.preprocessing import image
import numpy as np

# Function to preprocess the image for prediction
def preprocess_image(img_path):
    img = image.load_img(img_path, target_size=(img_size, img_size))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array /= 255.0  # Rescale the image
    return img_array

# Function to predict the fruit class
def predict_fruit_class(img_path):
    img = preprocess_image(img_path)
    predictions = model.predict(img)
    predicted_class = np.argmax(predictions, axis=1)
    class_labels = list(train_generator.class_indices.keys())
    predicted_fruit = class_labels[predicted_class[0]]
    return predicted_fruit

# Function to get nutritional values for the predicted fruit
def get_nutritional_values(fruit_name):
    if fruit_name in nutritional_info:
        return nutritional_info[fruit_name]
    else:
        return "Nutritional data not available for this fruit."

# Example usage:
img_path = '/content/fruits-360_dataset_100x100/fruits-360/Test/Banana 1/102_100.jpg'  # Replace with your image path
predicted_fruit = predict_fruit_class(img_path)
print(f"Predicted fruit: {predicted_fruit}")

# Get nutritional values
nutrition = get_nutritional_values(predicted_fruit)
print(f"Nutritional Information for {predicted_fruit}: {nutrition}")
```
